```python
import cv2
```

## 이미지 읽기
### cv2.imread(filename, flag)

1. filename은 절대/상대 경로

2. flag   
 cv2.IMREAD_COLOR (Default) : 1   
 cv2.IMREAD_GRAYSCALE : 0    
 cv2.IMREAD_UNCHANGED : -1   

```python
original = cv2.imread('image.jpg', cv2.IMREAD_COLOR)
gray = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
unchange = cv2.imread('image.jpg', cv2.IMREAD_UNCHANGED)
# return : array
```

```python
original.shape
# (행, 열, 채널 수(BGR)))
```

---
## 이미지 보기
### cv2.imshow(title, image)

```python
cv2.imshow("Original",original)
cv2.imshow("Gray",gray)
cv2.imshow("Unchange",unchange)
cv2.waitKey(0)
# 키보드 입력을 기다림 0은 무한 대기, o.w 입력 시간만큼
cv2.destroyAllWindows()
# 화면 종료
```

---
## 이미지 저장하기
### cv2.imwrite(filename, image)

```python
# esc를 누르면 종료, s를 누르면 저장
cv2.imshow("Gray",gray)
key = cv2.waitKey(0)
if key == 27: # keyboard input = esc
    cv2.destroyAllWindows()
elif key == ord('s'): # input = ' s'
    cv2.imwrite("GRAY_IMAGE.png",gray)
    cv2.destroyAllWindows()
```

## Matplotlib 사용하기

```python
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg', 1)

plt.imshow(img)
plt.xticks([]) # 눈금을 없애기 위함
plt.yticks([])
plt.show()
```

OpenCV는 BGR 순서로 읽는 반면, Matplotlib은 RGB 순서로 읽기 때문에,
색이 뒤집혀있음

```python
b, g, r = cv2.split(img)   # img파일을 b,g,r로 분리
img2 = cv2.merge([r,g,b]) # r,g,b로 순서를 정렬하여 Merge하였음

plt.imshow(img2)
plt.xticks([])
plt.yticks([]) 
plt.show()
```

색이 정상적으로 나오는 것을 확인
