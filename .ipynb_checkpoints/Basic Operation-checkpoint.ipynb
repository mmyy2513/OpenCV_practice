```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

---
## Pixel Value

```python
img = cv2.imread('image.jpg')
b,g,r = cv2.split(img)
img = cv2.merge([r,g,b])

# col:100, row:200에 red circle
img = cv2.circle(img, (100,200), 10, (255,0,0), -1)

# pixel값에 접근할 때는 (row, col)로 접근
px = img[100,200]
print("row:100, col:200\npixel value :",px)

px_r = img[200,100]
print("\nrow:200, col:100")
print("pixel value :",px_r)

plt.imshow(img)
plt.show()
```

```python
# 각 채널의 pixel 값은 img[row,col,0/1/2](r/g/b)
# 원래 b/g/r이나 r/g/b로 정렬되어 있기 때문에
print(img[100,200])
print(img[100,200,0])
print(img[100,200,1])
print(img[100,200,2])
```

```python
for i in range(0,img.shape[0]):
    img[i,200] = [0,0,0]
plt.imshow(img)
plt.show()
```

```python
# numpy를 이용하여 다음과 같이 변경
for i in range(img.shape[1]):
    # img.item(row,col,rgb) : 해당 픽셀 rgb value 확인
    # img.itemset((row,col,rgb), value) : 해당 픽셀 rgb 값을 원하는 value로
    img.itemset((400,i,0),0) 
plt.imshow(img)
plt.show()
```

## 이미지 ROI

```python
# img[row의 시작:끝, col의 시작:끝]
head = img[150:500, 700:1400]
for i in range(3):
    img[350*i:350+350*i, 0:700] = head
plt.imshow(img)
plt.show()
```

## 이미지 Channels

```python
# 다음과 같이 cv2.split을 이용하여 채널 분리 가능
img = cv2.imread("image.jpg")
b, g, r = cv2.split(img)
```

```python
# indexing을 이용하여 채널 분리
b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]
```

cv2.split은 비용이 많이 들기 때문에, indexing 방법이 보다 효율적

```python
dst = img.copy()
img[:,:,2] = 0 # red channel의 값을 0으로

cv2.imshow("origin",dst)
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
