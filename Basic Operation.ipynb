```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

---
## Pixel Value

```python
img = cv2.imread('image.jpg')
b,g,r = cv2.split(img)
img = cv2.merge([r,g,b])

# col:100, row:200 red circle
img = cv2.circle(img, (100,200), 10, (255,0,0), -1)

# pixel   (row, col) 
px = img[100,200]
print("row:100, col:200\npixel value :",px)

px_r = img[200,100]
print("\nrow:200, col:100")
print("pixel value :",px_r)

plt.imshow(img)
plt.show()
```

```python
#   pixel  img[row,col,0/1/2](r/g/b)
#  b/g/r r/g/b   
print(img[100,200])
print(img[100,200,0])
print(img[100,200,1])
print(img[100,200,2])
```

```python
for i in range(0,img.shape[0]):
    img[i,200] = [0,0,0]
plt.imshow(img)
plt.show()
```

```python
# numpy    
for i in range(img.shape[1]):
    # img.item(row,col,rgb) :   rgb value 
    # img.itemset((row,col,rgb), value) :   rgb   value
    img.itemset((400,i,0),0) 
plt.imshow(img)
plt.show()
```
