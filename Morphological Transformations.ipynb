```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Erosion

- 작은 object 제거에 효과적
### cv2.erode(src, kernel, dst, anchor, iterations, borderType, borderValue)
---

## Dilation

- 경계가 부드러워지고, 구멍이 메꿔지는 효과
### cv2.dilation(src, kernel, dst, anchor, iterations, borderType, borderValue)
---

## Opening & Closing
- Opening : Erosion 후, Dilation 적용 : 작은 object 제거에 적합
- Closing : 전체적인 윤곽 파악에 적합
### cv2.morphologyEx(src, op, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]])
- op
    - MORPH_OPEN
    - MORPH_CLOSE
    - MORPH_GRADIENT : Dilation과 Erosion의 차이
    - MORPH_TOPHAT : Opening과 원본의 차이
    - MORPH_BLACKHAT : Dilation과 원본의 차이
    
    
---

## Structing Element
: 커널을 만듦
### cv2.getStructingElement(shape, ksize, [,anchor])
- shape
    - MORPH_RECT : 사각형
    - MORPH_ELLIPSE : 타원
    - MORPH_CROSS : 십자
---

```python
img = cv2.imread("src/noisy2.jpg",1)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))

erosion = cv2.erode(img, kernel, iterations = 1)
dilation = cv2.dilate(img, kernel, iterations = 1)
opening1 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
opening2 = cv2.morphologyEx(opening1, cv2.MORPH_OPEN, kernel)
opening3 = cv2.morphologyEx(opening2, cv2.MORPH_OPEN, kernel)

closing1 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
closing2 = cv2.morphologyEx(closing1, cv2.MORPH_CLOSE, kernel)
closing3 = cv2.morphologyEx(closing2, cv2.MORPH_CLOSE, kernel)

imglist = [img,erosion, dilation, opening1,opening3, closing1,closing3]
titles = ["original","Erosion",'Dilation',"Opening1",'Opening3','Closing1','Closing3']
plt.figure( figsize = (10,20))
for i,image in enumerate(imglist):
    plt.subplot(4,2,i+2)
    plt.title(titles[i])
    plt.xticks([]);plt.yticks([])
    plt.imshow(image)
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))

erosion = cv2.erode(img, kernel, iterations = 1)
dilation = cv2.dilate(img, kernel, iterations = 1)
opening1 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
opening2 = cv2.morphologyEx(opening1, cv2.MORPH_OPEN, kernel)
opening3 = cv2.morphologyEx(opening2, cv2.MORPH_OPEN, kernel)

closing1 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
closing2 = cv2.morphologyEx(closing1, cv2.MORPH_CLOSE, kernel)
closing3 = cv2.morphologyEx(closing2, cv2.MORPH_CLOSE, kernel)

imglist = [img,erosion, dilation, opening1,opening3, closing1,closing3]
titles = ["original","Erosion",'Dilation',"Opening1",'Opening3','Closing1','Closing3']
plt.figure( figsize = (10,20))
for i,image in enumerate(imglist):
    plt.subplot(4,2,i+2)
    plt.title(titles[i])
    plt.xticks([]);plt.yticks([])
    plt.imshow(image)
plt.show()
```
