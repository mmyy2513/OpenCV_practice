---
## Line 그리기
### cv2.line(img, start, end, color, thickness)
- img : 그림을 그릴 파일
- start : 시작 좌표
- end : 끝 좌표
- color : RGB
- thickness : 두께

```python
import numpy as np
import matplotlib.pyplot as plt

# 검은 512x512 캔버스
img = np.zeros((512,512,3),np.uint8)
# (0,0) - (512,512) 흰 직선
img = cv2.line(img, (0,0), (512,512), (255,255,255), 2)

"""
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
"""

plt.imshow(img)
plt.show()
```

---
## 사각형 그리기
### cv2.rectangle(img, start, end, color, thickness)
- start : left-top 좌표
- end : right-bottom 좌표

```python
img = cv2.rectangle(img, (100,100), (400,400), (255,0,0), 2)
plt.imshow(img)
plt.show()
```

---
## 원 그리기
### cv2.circle(img, center, radian, color, thickness)
- thickness : -1이면 안쪽을 채움

```python
img = cv2.circle(img, (250,250), 150, (0,0,255), 2)
plt.imshow(img)
plt.show()
```

---
## 타원 그리기
### cv2.ellipse(img, ((center), (box), (degree)), color, thickness)
- center : 타원 중심
- box : 타원이 채울 박스 좌표
- degree : 회전 각도 ( 시계 반대 방향 )

```python
img = cv2.ellipse(img, ((250,250),(400,150),45), (0,255,0), 2)
plt.imshow(img)
plt.show()
```

---
## Polygon 그리기
### cv2.polylines(img, pts, isClosed, color, thickness)
- pts : 꼭지점 좌표
- 닫힌 도형 여부

```python
pts = np.array([[250,100],[100,250],[400,250],[200,400],[300,400]],np.int32)
img = cv2.polylines(img, [pts], True, (255,0,255), 2)
plt.imshow(img)
plt.show()
```

### cv2.fillConvexPoly(img, pts, color)
- 채워진 다각형, 두께가 필요 없다

```python
pts = np.array([[250,100],[100,250],[170,400],[330,400],[400,250]],np.int32)
img = cv2.fillConvexPoly(img, pts, (255,0,255))
plt.imshow(img)
plt.show()
```

## 이미지에 Text 추가
### cv2.putText(img, text, org, font, fontScale, color)
- text : 문자열
- org : 문자열 left-bottom 좌표
- font : 폰트 타입 ( cv2.FONT_???)
- fontScale : 폰트 사이즈


```python
cv2.putText(img, "minyoung", (100,480), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,255,255))
plt.imshow(img)
plt.show()
```
